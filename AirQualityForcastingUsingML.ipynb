```python
air_quality_data = air_quality_data.replace(-200, np.nan)
```

```python
#replacing the missing values to the mean values
```

```python
cols = ['CO(GT)', 'NO2(GT)', 'T', 'RH']  # example columns
air_quality_data[cols] = air_quality_data[cols].fillna(air_quality_data[cols].mean())
```

 **Forcasting with FB Prophet Algorithm**

Link: https://facebook.github.io/prophet/docs/quick_start.html#python-api

```python
#converting the date format from dd/mm/yyyy to yyyy/mm/dd
date_info= pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
print(date_info)
```

```python
time_info= air_quality_data['Time']
```

```python
print(time_info)
```

```python
# Suppose your column name is "Time"
time_info = air_quality_data['Time']

# Replace '.' with ':' in each value
time_info = time_info.apply(lambda x: x.replace('.', ':'))

# Optional: update the DataFrame
air_quality_data['Time'] = time_info
```

```python
#combining two series to a panda dataframe
date_time = pd.concat([date_info, air_quality_data['Time']], axis=1)
```

```python
data = pd.DataFrame()
data['ds'] = pd.to_datetime(date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str))
```

```python
data.head()
print(type('data'))
```

```python
data['y'] = air_quality_data['RH']
print(data)
```

```python
from prophet import Prophet
```

```python
#training the prophet model
model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```
